# 函式 function

須將一些重複或有特別定義的程式，拆分成容易管理的小程式，這些小程式就稱為「函式」

```python
#無參數
#定義一個名為hello的函數，並呼叫它

def hello():
  print('hello')
hello()
```

## 函式回傳值

函式除了可以傳入參數，也可以使用「return」回傳程式運算後的
結果，回傳的結果不限型態，可以是數字、字串、串列、tuple...等

```python
def a(x,y):
  result=x+y*2
  return result

b=a(1,2)
c=a(2,3)

print(b)
print(c)
```

## while 迴圈

「while 迴圈」是「根據條件判斷，決定是否重複或停止」的迴圈，用法為「while 條件:」，如果條件判斷為 True，就會不斷執行迴圈內容，如果判斷為 False，就會停止迴圈

### 解釋以下內容為何 b 跟 c 結果不一樣:

B 將函式 a 假設 x 為1進去跑迴圈，當result=5時，它會成立"if result==5:"，因此執行return，，中斷迴圈，結果為5。

C 將函式 a 假設 x 為2進去跑迴圈，當result=10時，它會不符合迴圈條件"while result<10:"，因此執行return，，中斷迴圈，結果為10。


```python
def a(x):
  result=0
  while result<10:
    result=result+x
    if result==5:
      return result
  return result

b=a(1)
c=a(2)

print(b)
print(c)
```

### 「return」回傳運算後的結果

```python
def tf(value):
  return 2 * value - 1

# 測試這個函數
test_values = [0.0, 0.0011, 0.25, 0.5,0,54, 0.75, 1.0]
scaled_values = [tf(value) for value in test_values]
print("Original values: ", test_values)
print("Scaled values: ", scaled_values)
```

## 遞迴 recursion

• 函式會自己呼叫自己。

• 具備函式停止條件 ( 避免無窮盡的呼叫自己 )。

```python
def a(n):
  if n==0 or n==1:
    return 1
  else:
    return n+a(n-1)

print(a(3))
```

遞迴過程:

a(0) - 1

a(1) - 1

a(2) - 2+a(1)=3

a(3) - 3+a(2)=6

a(4) - 4+a(3)=10

a(5) - 5+a(4)=15
