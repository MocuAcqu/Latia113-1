## 作業-改寫輸出成CSV檔

## 解析

```python
soup = BeautifulSoup(html, "html.parser") # create a new bs4 object from the html data loaded
for script in soup(["script", "style"]): # remove all javascript and stylesheet code
  script.extract()
  # get text
  text = soup.get_text()
  # break into lines and remove leading and trailing space on each
  lines = (line.strip() for line in text.splitlines())
  # break multi-headlines into a line each
  chunks = (phrase.strip() for line in lines for phrase in line.split(" "))
  # drop blank lines
  texts = '\n'.join(chunk for chunk in chunks if chunk)
print(texts)
```

# 作業-downloadcsv

溫室氣體排放（工業與家庭）：2024 年 6 月季度 – CSV

```python
import requests
url = 'https://www.stats.govt.nz/assets/Uploads/Greenhouse-gas-emissions-industry-and-household/Greenhouse-gas-emissions-industry-and-household-June-2024-quarter/Download-data/greenhouse-gas-emissions-industry-and-household-June-2024-quarter.csv' # 牌告匯率 CSV 網址
rate = requests.get(url) # 爬取網址內容
rate.encoding = 'utf-8' # 調整回應訊息編碼為 utf-8，避免編碼不同造成亂碼
rt = rate.text # 以文字模式讀取內容
rts = rt.split('\n') # 使用「換行」將內容拆分成串列
for i in rts: # 讀取串列的每個項目
  try: # 使用 try 避開最後一行的空白行
    a = i.split(',') # 每個項目用逗號拆分成子串列
    print(a[3] + ': ' + a[4]) # 取出第一個 ( 0 ) 和第十三個項目 ( 12 )
  except:
    break
```

```python
import requests
import matplotlib.pyplot as plt
from datetime import datetime
from collections import defaultdict

# 爬取資料
url = 'https://www.stats.govt.nz/assets/Uploads/Greenhouse-gas-emissions-industry-and-household/Greenhouse-gas-emissions-industry-and-household-June-2024-quarter/Download-data/greenhouse-gas-emissions-industry-and-household-June-2024-quarter.csv'
rate = requests.get(url)  # 爬取網址內容
rate.encoding = 'utf-8'  # 調整編碼為 utf-8，避免亂碼
rt = rate.text  # 以文字模式讀取內容
rts = rt.split('\n')  # 使用換行符分割內容為多行

# 初始化數據儲存
data_by_category = defaultdict(list)  # 使用字典以 Anzsic_descriptor 分組數據

# 逐行處理資料
for i in rts:
    try:
        a = i.split(',')  # 用逗號拆分每行
        period = a[3]  # 第 4 列為 Period
        value = a[4]  # 第 5 列為 Data_value
        category = a[1]  # 第 2 列為 Anzsic_descriptor

        # 過濾空值並轉換數據類型
        if period and value and category:
            # 將 "年.月" 轉為 datetime 格式
            year, month = period.split('.')
            period_dt = datetime(int(year), int(month), 1)  # 將年和月轉為 datetime

            # 儲存分類數據
            data_by_category[category].append((period_dt, float(value)))
    except (IndexError, ValueError):
        # 避免最後一行為空或資料不完整導致錯誤
        continue

# 為每個分類生成一張折線圖
for category, data in data_by_category.items():
    # 按時間排序
    sorted_data = sorted(data, key=lambda x: x[0])
    periods, values = zip(*sorted_data)  # 解壓為時間和數值列表

    # 繪製折線圖
    plt.figure(figsize=(12, 6))
    plt.plot(periods, values, marker='o', linestyle='-', color='b')

    # 圖表美化
    plt.title(f'Greenhouse Gas Emissions for {category}', fontsize=16)
    plt.xlabel('Time (Year-Month)', fontsize=12)
    plt.ylabel('Emissions (Data Value)', fontsize=12)
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()

    # 保存圖表（可選，使用 plt.savefig）
    # plt.savefig(f'{category}_emissions.png')

    # 顯示圖表
    plt.show()
```

     - 自2021年以來，排放量部分回升，但未恢復到2019年的最高水平。
   - **結論：**
     - 建築活動在經濟復甦後仍未完全回到疫情前的活躍程度，可能反映出疫情後行業需求的轉變。

#### 2. **Manufacturing 部門**

     - 從2010年到2018年，Manufacturing（製造業）部門的排放量穩步上升，顯示出行業擴張。
     - 2019年以後，排放量開始下降，並在2020年之後顯著減少。
   - **結論：**
     - 排放量的下降可能與行業轉型或全球對減少碳排放的政策有關。
     - 可能有更多工廠引入了綠色技術，或是製造規模縮小。

#### 3. **Mining 部門**

     - 2020年疫情期間進一步下降，且後續波動性較大。
   - **結論：**
     - 礦業的排放量整體呈下降趨勢，這可能與行業萎縮或轉向低碳能源開採有關。
     - 短期波動可能反映了國際礦產品需求的變化。
